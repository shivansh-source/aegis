```python
# Cell 1: Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# For nicer inline plots
%matplotlib inline
```

```python
# Cell 2: Load the dataset
df = pd.read_csv(r'C:\Users\Srijit\OneDrive\Desktop\ai-in-defense\ai-model\dataset\war_history.csv') # "./" means current directory
df.head()
```

```python
# Cell 3: Basic info and null checks
print(df.info())
print(df.isnull().sum())
```

```python
# Cell 4: Summary statistics
df.describe().T
```

```python
# Cell 4: Summary statistics
df.describe().T
```

```python
# Cell 5: Distribution of troop counts
plt.figure()
plt.hist(df['troopsA'], bins=30)
plt.title('Troops A Distribution')
plt.xlabel('Number of Troops')
plt.ylabel('Frequency')
plt.show()

plt.figure()
plt.hist(df['troopsB'], bins=30)
plt.title('Troops B Distribution')
plt.xlabel('Number of Troops')
plt.ylabel('Frequency')
plt.show()
```

```python
# Cell 6: Distribution of modernization levels
plt.figure()
plt.hist(df['modA'], bins=20)
plt.title('Weapon Modernization A')
plt.xlabel('Modernization (%)')
plt.ylabel('Frequency')
plt.show()

plt.figure()
plt.hist(df['modB'], bins=20)
plt.title('Weapon Modernization B')
plt.xlabel('Modernization (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Cell 7: Correlation heatmap
corr = df[['troopsA','troopsB','modA','modB']].corr()
plt.figure(figsize=(6,5))
plt.imshow(corr, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation=45)
plt.yticks(range(len(corr)), corr.columns)
plt.title('Feature Correlation')
plt.show()
```

```python
# Cell 8: Feature engineering  compute strength and ratio
df['strengthA'] = df['troopsA'] * df['modA']
df['strengthB'] = df['troopsB'] * df['modB']
df['strength_ratio'] = df['strengthA'] / (df['strengthA'] + df['strengthB'])
df[['strengthA','strengthB','strength_ratio']].head()
```

```python
# Cell 9: Visualize strength ratio distribution
plt.figure()
plt.hist(df['strength_ratio'], bins=30)
plt.title('Strength Ratio Distribution (A / (A + B))')
plt.xlabel('Strength Ratio')
plt.ylabel('Frequency')
plt.show()
```

```python
# Cell 10: Strength ratio vs. observed win rate
df['A_won'] = (df['winner'] == df['countryA']).astype(int)
# Bin into deciles
df['ratio_bin'] = pd.cut(df['strength_ratio'], bins=np.linspace(0,1,11))
win_rates = df.groupby('ratio_bin')['A_won'].mean().reset_index()
plt.figure(figsize=(8,4))
plt.plot(win_rates['ratio_bin'].astype(str), win_rates['A_won'], marker='o')
plt.xticks(rotation=45)
plt.title('Observed Win Rate of A by Strength Ratio Decile')
plt.xlabel('Strength Ratio Bin')
plt.ylabel('Win Rate of A')
plt.show()
```

```python
# Cell 11: GDP loss relationship with strength
plt.figure()
plt.scatter(df['strengthA'], df['gdp_loss_A'], alpha=0.3, label='A')
plt.scatter(df['strengthB'], df['gdp_loss_B'], alpha=0.3, label='B', color='red')
plt.legend()
plt.title('Strength vs. GDP Loss')
plt.xlabel('Strength')
plt.ylabel('GDP Loss')
plt.show()
```
